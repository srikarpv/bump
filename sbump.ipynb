```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import style 
from math import *


style.use('ggplot')

xaxis, x, y, z, a = np.loadtxt('data.csv', delimiter = ',', unpack = True)

def graph():
    plt.plot(xaxis,x)
    plt.plot(xaxis,y)
    plt.plot(xaxis,z)
    plt.plot(xaxis,a)


    plt.title('Accelerometer Data ')
    plt.ylabel('')
    plt.xlabel('readings')

    plt.show()

graph()

```
